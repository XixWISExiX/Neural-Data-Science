# General Linear Model

## Notation

```python
# Step 1: Generate a random set of one dependent and one independent variable with 30 observations and 0.8 correlation.
# Make sure the shape of the returned variable is (30, 1) not (30,).
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
cov_mat=np.array([[1,.8],[.8,1]])
mean_vec=[3,7]
data=np.random.multivariate_normal(mean=mean_vec,cov=cov_mat,size=30)
X=data[:,0,np.newaxis]
y=data[:,1,np.newaxis]
```

## The Intercept

```python
# Step 2: Create the Design Matrix (with intercept)
intercept=np.ones((y.size,1))
tuples_with_arrays=(intercept,X)
x_icp=np.hstack(tuples_with_arrays)
print(x_icp)

# Plot the data
plt.figure(figsize=(5, 5))
plt.scatter(x_icp[:,1],y)
plt.xlabel('X', fontsize=25)
plt.ylabel('y', fontsize=25)
plt.show()
```

## Estimating Parameters

```python
# Step 3: Apply OLS to estimate the parameters.
from numpy.linalg import inv
est_betas=inv(x_icp.T @ x_icp)@x_icp.T@y
print(est_betas)
```

## Predicting Values

```python
# Step 4: Calculate the predicted values and plot them.
y_hat=x_icp @ est_betas
print(y_hat)
plt.figure(figsize=(5,5))
plt.scatter(X,y)
plt.xlabel('X',fontsize=25)
plt.ylabel('Y',fontsize=25)
plt.scatter(X,y_hat,c='orange')
plt.legend(['True Y','Predicted y'])
plt.grid()
plt.show()
```

## Residuals and Model Fit

```python
# Step 5: Calculate the residuals and plot them.
MSE=np.sum((y-y_hat) **2) /len(y)
print(MSE)
```

```python
# Calculate the R-squared value.
r_squared=1-(np.sum((y-y_hat) **2))/(np.sum((y-np.mean(y)) **2))
print(r_squared)
```
